**Step 4: Data Preprocessing**

```python
# Drop Irrelevant Feature
housing_df.drop(columns=['No.'], inplace=True)

# Feature Engineering: Creating Derived Features
housing_df['rooms_per_household'] = housing_df['total_rooms'] / housing_df['households']
housing_df['bedrooms_per_room'] = housing_df['total_bedrooms'] / housing_df['total_rooms']
housing_df['population_per_household'] = housing_df['population'] / housing_df['households']

# Fill missing values in 'bedrooms_per_room' with the median
housing_df['bedrooms_per_room'].fillna(housing_df['bedrooms_per_room'].median(), inplace=True)
# One-Hot Encoding for 'ocean_proximity'
housing_df = pd.get_dummies(housing_df, columns=['ocean_proximity'], drop_first=True)

# Keep Only the Most Relevant Categorical Feature
if 'ocean_proximity_INLAND' in housing_df.columns:
    categorical_features = ['ocean_proximity_INLAND']
    housing_df = housing_df[categorical_features + [col for col in housing_df.columns if col not in ['ocean_proximity_NEAR BAY', 'ocean_proximity_NEAR OCEAN', 'ocean_proximity_ISLAND']]]
else:
    print("Error: ocean_proximity_INLAND not found in dataset after encoding!")

# Drop Redundant Features
housing_df.drop(columns=['total_rooms', 'total_bedrooms', 'population', 'households'], inplace=True)

# Display the cleaned dataset
housing_df.head()
```

```python
# Boxplot for 'housing_median_age'
plt.figure(figsize=(8, 6))
plt.boxplot(housing_df['housing_median_age'], vert=False)
plt.title('Box Plot of Housing Median Age')
plt.xlabel('Housing Median Age')
plt.show()

# Boxplot for 'latitude' & 'total_rooms'
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot(housing_df['latitude'], vert=False)
plt.title('Box Plot of Latitude')

plt.subplot(1, 2, 2)
plt.boxplot(housing_df['rooms_per_household'], vert=False)
plt.title('Box Plot of Rooms Per Household')

plt.show()
```

```python
# Calculate Q1, Q3, and IQR for 'rooms_per_household'
Q1 = housing_df['rooms_per_household'].quantile(0.25)
Q3 = housing_df['rooms_per_household'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
housing_df = housing_df[~((housing_df['rooms_per_household'] < lower_bound) | (housing_df['rooms_per_household'] > upper_bound))]
```

**XGBRegressor(Main Model)**

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Feature Scaling (optional for XGBoost but can help performance)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the XGBoost Regressor model with L1 (reg_alpha) and L2 (reg_lambda) regularization
xgb_model = XGBRegressor(
    n_estimators=1000,  # Number of boosting rounds
    learning_rate=0.05,  # Step size shrinkage
    max_depth=6,  # Tree depth
    subsample=0.8,  # Subsample ratio of training instances
    colsample_bytree=0.8,  # Subsample ratio of columns per tree
    reg_alpha=0.1,  # L1 Regularization (Lasso) - Adjust as needed
    reg_lambda=1.0,  # L2 Regularization (Ridge) - Adjust as needed
    random_state=42
)

xgb_model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = xgb_model.predict(X_test_scaled)

# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"XGBoost Model - MSE: {mse:.2f}, R Score: {r2:.4f}")

# Plot actual vs predicted values
plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred, alpha=0.5, label="Predicted vs Actual")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='dashed')  # 45-degree line
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("XGBoost Predictions vs Actual")
plt.legend()
plt.show()
```

```python
# prompt: can you make complex hyper tunning for XGBoost

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# Define the parameter grid for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(50, 1500),
    'learning_rate': uniform(0.01, 0.3),
    'max_depth': randint(3, 10),
    'subsample': uniform(0.5, 0.5),
    'colsample_bytree': uniform(0.5, 0.5),
    'reg_alpha': uniform(0, 1),
    'reg_lambda': uniform(0, 1),
}

# Create the XGBoost Regressor model
xgb_model = XGBRegressor(random_state=42, objective='reg:squarederror')  # Use appropriate objective function

# Create RandomizedSearchCV object
random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=param_dist,
    n_iter=50,  # Number of parameter settings sampled
    scoring='neg_mean_squared_error',  # Use an appropriate scoring metric
    n_jobs=-1,  # Use all available CPU cores
    cv=5,  # Number of cross-validation folds
    verbose=2,
    random_state=42
)

# Fit the RandomizedSearchCV object to the training data
random_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters and model
best_params = random_search.best_params_
best_xgb_model = random_search.best_estimator_

print("Best Hyperparameters:", best_params)

# Evaluate the best model on the test set
y_pred = best_xgb_model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Best XGBoost Model - MSE: {mse:.2f}, R Score: {r2:.4f}")
```

```python
# prompt: handle missing value Pclass, SibSp, Parch, Fare, Survived

import pandas as pd
# Handle missing values for 'Pclass', 'SibSp', 'Parch', 'Fare', and 'Survived'
for col in ['Pclass', 'SibSp', 'Parch', 'Fare', 'Survived']:
    if df[col].isnull().any():
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col].fillna(df[col].median(), inplace=True)  # Fill with median for numeric columns
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)  # Fill with mode for non-numeric columns
```

```python

import matplotlib.pyplot as plt
import numpy as np
# Assuming 'df' is your DataFrame from the previous code

# Identify outliers using the IQR method for numerical features
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']
for col in numerical_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in '{col}':")
    print(outliers)

    # Optionally, replace outliers with the bounds
    df[col] = np.clip(df[col], lower_bound, upper_bound)
    # Or remove outliers
    #df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

# You can also visualize the outliers using box plots:
for col in numerical_features:
  plt.figure(figsize=(8, 6))
  sns.boxplot(x=df[col])
  plt.title(f'Box Plot of {col}')
  plt.show()
```
